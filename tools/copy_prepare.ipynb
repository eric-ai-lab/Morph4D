```python
import os, os.path as osp
import shutil

src = "../data/iphone_1x_dev"

dst = "../data/iphone_final"

for scene in os.listdir(src):
    print(scene)
    src_scene = osp.join(src, scene)
    dst_scene = osp.join(dst, scene)
    os.makedirs(dst_scene, exist_ok=True)
    for fn in os.listdir(src_scene):
        if fn == "log":
            continue
        cmd = f"cp -r {osp.join(src_scene, fn)} {osp.join(dst_scene, fn)}"
        print(cmd)
        os.system(cmd)
```
